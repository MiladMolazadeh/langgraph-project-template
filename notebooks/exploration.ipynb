## Run the Agent
> Requires a valid API key for the configured provider in `.env`

```python
from langchain_core.messages import HumanMessage
from my_agent.utils.tracing import get_callbacks

try:
    result = graph.invoke(
        {'messages': [HumanMessage(content='What is 42 * 17?')]},
        config={
            'configurable': {'thread_id': 'notebook-1'},
            'callbacks': get_callbacks(),
        },
    )
    for msg in result['messages']:
        print(f'{msg.__class__.__name__}: {msg.content[:300]}')
        print('---')
except Exception as e:
    print(f'Invocation failed (check API keys in .env): {type(e).__name__}: {e}')
```
